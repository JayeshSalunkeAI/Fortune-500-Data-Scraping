```python
# Required Libraries
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
```

```python
base_url = 'https://fortune.com/ranking/global500/'
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}

main_responce = requests.get(base_url, headers=header)
main_responce.status_code
```

```python
main_html = main_responce.text
main_soup = BeautifulSoup(main_html, 'html.parser')
```

```python
chrome_options = Options()
chrome_options.page_load_strategy = 'normal'
driver = webdriver.Chrome(options=chrome_options)
driver.set_page_load_timeout(60)

driver.get(base_url)
time.sleep(3)
html = driver.page_source
soup_main = BeautifulSoup(html, "html.parser")
driver.quit()
```

```python
tr_container = soup_main.find_all('td', attrs={'data-cy':'list-cell'})
```

```python
tr_container
```

```python
data = []
for i in range(len(tr_container)):
    info = tr_container[i].get_text().strip()                  
    data.append(info)
data
```

```python
# Define your columns
columns = ['Rank', 'Company', 'Revenue', 'Growth', 'Profit', 'Profit_Margin', 
           'Assets', 'Employees', 'Change', 'Years']

# Reshape the flat list into rows (assuming 10 columns per row)
rows = []
for i in range(0, len(data), 10):
    row = data[i:i+10]
    if len(row) == 10:  # Only add complete rows
        rows.append(row)

# Create DataFrame
df = pd.DataFrame(rows, columns=columns)
```

```python
df.to_csv("Fortune500.csv")
```
